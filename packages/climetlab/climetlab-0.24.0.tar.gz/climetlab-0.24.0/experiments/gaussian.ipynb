```python
import climetlab as cml
import matplotlib.pyplot as plt

from climetlab.grids import unstructed_to_structed

ds = cml.load_source("mars", param="2t", date=20220907, levtype="sfc")
```

```python
unstructed_to_structed(ds[0])  # Load caches

x = list(range(8, 20, 1))
y = [unstructed_to_structed(ds[0], chunk_size=c) for c in x]
plt.plot(x, y)
```
