```python
import climetlab as cml
```

```python
source = cml.load_dataset("sample-bufr-data")
```

```python
pd = source.to_pandas(
    columns=(
        "stationNumber",
        "latitude",
        "longitude",
        "data_datetime",
        "pressure",
        "airTemperature",
    ),
    filters={},
)
```
